```python
## Importar librería

import pandas as pd
import warnings
from sklearn.preprocessing import StandardScaler
warnings.filterwarnings('ignore', category=DeprecationWarning) 
```

```python
## Cargar datos 

data = pd.read_csv('../data/processed/data_eda.csv').sample(frac=0.5)
data.head()
```

```python
X = data.drop('CSAT Score',axis=1)
y = data['CSAT Score']
```

```python
y.isnull().sum()
```

```python
## Normalización
scaler = StandardScaler()
df_scaler = scaler.fit_transform(X)
data_scaler_X = pd.DataFrame(df_scaler, columns=X.columns)
data_scaler_X ['CSAT Score']=y.values
```

```python
data_scaler_X.to_csv('../data/processed/data_normalized.csv', index=False)
```
